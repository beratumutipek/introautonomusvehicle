**Tasks:**

````python
# generating an L-shaped point cloud.
import numpy as np
np.random.seed(0) # For reproducable results
points = np.hstack([np.array([[5,0.5]]).T* np.random.rand(2,100),np.array([[0.5,5]]).T*np.random.rand(2,100)])
print('Point cloud dimension:',points.shape)
````

````python
# Plot points here
import matplotlib.pyplot as plt
x = points[0, :]
y = points[1, :]
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Point Cloud')
plt.show()
````

````python
# write angle2rotMat(...):
"""
Create a 2D rotation matrix for a given angle in degrees.

    Input:
    angle (float): The angle in degrees for rotation.

    Returns:
    The 2x2 rotation matrix that represents the rotation by the specified angle.
    
    This function takes an angle in degrees and returns a 2D rotation matrix that represents
    a rotation by that angle. The input angle is converted to radians internally for
    the rotation calculation.
    """
import numpy as np
def angle2rotMat(angle):
    angle_rad = np.deg2rad(angle)
    cos=np.cos(angle_rad)
    sin = np.sin(angle)
    rotation_mat = np.array([[cos,-sin], [sin, cos]])
    return rotation_mat
````

````python
# create rotation matrices
angle_R1 = 30
R1 = angle2rotMat(angle_R1)
angle_R2 = 60
R2 = angle2rotMat(angle_R2)
print(f"Rotation Matrix R1 (30 degrees):\n {R1}\nRotation Matrix R2 (60 degrees): \n{R2}")
````

````python
# create translation vector
translation = np.array([3,4]).reshape(2,1)

# apply Transformations 
points_r1 = np.dot(R1, points)
points_r2 = np.dot(R2, points)
points_t = points_r2 + translation
````

````python
# create subplots and plot results
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].scatter(points[0], points[1])
axes[0, 0].set_title('Original')

axes[0, 1].scatter(points_r1[0], points_r1[1])
axes[0, 1].set_title('Rotated by R1')

axes[1, 0].scatter(points_r2[0], points_r2[1])
axes[1, 0].set_title('Rotated by R2')

axes[1, 1].scatter(points_t[0], points_t[1])
axes[1, 1].set_title('Translated by (3, 4)')

plt.tight_layout()
plt.show()
````

2. Plot the trajectory of the robot in world coordinates. **[2]**
    - The trajectory is computed by iteratively applying the measured odometry to the previous position:
$$ x_{i+1} = x_{i} + \delta x_{i},\qquad y_{i+1} = y_{i} + \delta y_{i}$$
3. Plot the trajectory of the robot in world coordinates from 07:30:00 to 8:00:00. **[1]**

````python
# 1.
odom_file = "odom.data"
odometry_data = []
with open(odom_file, "r") as file:
    for line in file:
        elements = line.strip().split(',')
        if len(elements) == 5:
            hours, minutes, seconds, dx, dy = map(float, elements)
            hours, minutes, seconds = int(hours), int(minutes), int(seconds)
            odometry_data.append({
                "time": (hours, minutes, seconds),
                "dx": dx,
                "dy": dy
            })
initial_position = odometry_data[0]
print("Initial Position:", initial_position)
````

````python
# 2.
dx, dy = odometry_data[0]["dx"], odometry_data[0]["dy"]
trajectory = [(dx, dy)]
i = 0
for data_point in odometry_data:
    dx, dy = data_point["dx"], data_point["dy"]
    x_prev, y_prev = trajectory[i]  # Get the previous position
    x_new = x_prev + dx
    y_new = y_prev + dy
    trajectory.append((x_new, y_new))
    i+=1
print(odometry_data[1])
````

````python
# 2.
sequence_of_commands = ['down', 'right', 'up', 'up', 'right', 'right', 'down', 'down', 'down', 'left']
commands = {
    "up": (0, -1), "down" : (0, 1), "left":(-1, 0), "right":(1,0)
}
robot_position = (0, 0)
for action in sequence_of_commands:

    delta = commands[action]
    new_position = (robot_position[0] + delta[0], robot_position[1] + delta[1])
    
    if 0 <= new_position[0] < 5 and 0 <= new_position[1] < 5:
        robot_position = new_position
    
print("Final position of the robot:", robot_position)
````
